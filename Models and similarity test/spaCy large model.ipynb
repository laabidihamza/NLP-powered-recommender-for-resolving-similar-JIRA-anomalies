```python
import pandas as pd
```

```python
merged_df = pd.read_csv('cleaned_data.csv')
merged_df
```

```python
import gensim
from gensim import models
from gensim.similarities import WmdSimilarity
import spacy
import os
import numpy as np
nlp = spacy.load('en_core_web_lg')
```

```python
# ref_sent = merged_df.loc[merged_df['index'] == 80 , 'Description'].iloc[0]
ref_sent = merged_df['Description'][15]
ref_sent
```

```python
ref_sent_vec = nlp(ref_sent)
ref_sent_vec
```

```python
merged_df['Description'] = merged_df['Description'].astype(str)
```

```python
all_docs = [nlp(row) for row in merged_df['Description']]
```

```python
sims = []
doc_id = []
for i in range(len(all_docs)):
    sim = all_docs[i].similarity(ref_sent_vec)
    sims.append(sim)
    doc_id.append(i)
    sims_docs = pd.DataFrame(list(zip(doc_id, sims)) , columns = ['doc_id', 'sims'])
```

```python
sims_docs_sorted = sims_docs.sort_values(by = 'sims', ascending = False)
```

```python
top5_sim_docs = merged_df.iloc[sims_docs_sorted['doc_id'][1:6]]
```

```python
ref_sent
```

```python
top_sim_scores = pd.concat([top5_sim_docs, sims_docs_sorted['sims']], axis = 1)
```

```python
for (text, scosimre) in zip(top_sim_scores['Description'], top_sim_scores['sims']):
    print(f"The top 5 Similarity sentences are: {text} \n with similarity score of {scosimre}")
```
